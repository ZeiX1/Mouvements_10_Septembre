## Code pour merge tous les datasets. A adapter selon l'environnement o√π vous avez stock√© les donn√©es. 

```python
parent_folder = (".")
# De mon c√¥t√©, j'ai enregistr√© mes csv dans des dossiers qui commencent tous par : scraped_messagesDDMM
folders = [f for f in os.listdir(parent_folder) if f.startswith("scraped_messages")] # je cr√©e une liste qui stocke tous les noms de mes dossiers. 

print(f"Dossiers d√©tect√©s : {folders}")

# Cr√©er le dossier de sortie
output_folder = os.path.join(parent_folder, "merged_messages_all_v1")
os.makedirs(output_folder, exist_ok=True)

# Lister tous les fichiers CSV pr√©sents dans au moins un dossier
all_files = set()
for folder in folders:
    files = [f for f in os.listdir(folder) if f.endswith(".csv")]
    all_files.update(files)

print(f"Nombre de fichiers : {len(all_files)}")

# Fusion des fichiers
for filename in all_files:
    dfs = []
    for folder in folders:
        file_path = os.path.join(folder, filename)
        if os.path.exists(file_path):
            df = pd.read_csv(file_path)
            df['source_folder'] = os.path.basename(folder)  # nom du dossier source
            dfs.append(df)
    
    merged = pd.concat(dfs, ignore_index=True)
    merged = merged.drop_duplicates(subset=["message_id", "date"])
    merged = merged.sort_values(by="date")

    output_file = os.path.join(output_folder, filename)
    merged.to_csv(output_file, index=False)
    print(f"Fichier fusionn√© : {output_file}")
```

```python

def descriptive_stats_clean(df):
    
    print("=== Aper√ßu du DataFrame ===")
    display(df.head(3))
    print("\nNombre de lignes :", len(df))
    print("Colonnes :", list(df.columns))
    print("-" * 80)
    
    summary = pd.DataFrame(columns=['Type', 'Nb valeurs uniques', 'Valeur la plus fr√©quente', 'Fr√©quence (%)', 'Nb de valeurs manquantes'])
    
    for col in df.columns:
        top_value = df[col].mode().iloc[0] if not df[col].mode().empty else None
        top_freq = df[col].value_counts(normalize=True).iloc[0] * 100 if not df[col].value_counts().empty else 0
        
        summary.loc[col] = [
            df[col].dtype,
            df[col].nunique(),
            str(top_value)[:80] + ("..." if len(str(top_value)) > 80 else ""),
            round(top_freq, 2),
            df[col].isna().sum()
        ]
    
    print("=== Statistiques descriptives ===")
    display(summary)

    print("\n=== Infos g√©n√©rales ===")
    df.info()
```

#### Premiere version de topic modeling
Cette premi√®re version utilise le topic modeling LDA. On doit transformer nos messages sous formes de listes de mots, puis cr√©er un dictionnaire avec la fr√©quence d'apparition du mot dans les messages. C'est la methode Bag of words(bow).

```python
from gensim.utils import simple_preprocess
from gensim.corpora import Dictionary
from gensim.models import CoherenceModel
from gensim.models import LdaMulticore
import gensim
```

```python
def preprocess(text):
    tokens = simple_preprocess(str(text),deacc= True)
    return [t for t in tokens if t not in stopwords] # On cr√©√© une liste de tokens(un token =~ un mot)
```

```python
# Fonction qui prend en argument un dataframe et un int(nombre de topic),lui applique la fonction preprocess au dessus puis utilise le package gensim pour g√©n√©rer les topics. 
# En output, on a le mod√®le entra√Æn√©, le corpus de mot( bag of words) et la coh√©rence du mod√®le, voir plus bas pour une d√©finition. 
def topic_model(df,nb_topic):
    df["token"] = df["text"].apply(preprocess)
    dictionary = Dictionary(df["token"])
    dictionary.filter_extremes(no_below=5, no_above=0.5)
    corpus = [dictionary.doc2bow(text) for text in df["token"]]
    lda = gensim.models.LdaMulticore(corpus, num_topics = nb_topic, id2word = dictionary, passes = 2) #  (6 c'est peut etre beaucoup)
    coherence_model_lda = CoherenceModel(model=lda, texts=df["token"], dictionary=dictionary)
    coherence_lda = coherence_model_lda.get_coherence()
    return (lda,corpus, coherence_lda, dictionary)
```

La coh√©rence(compris entre 0 et 1) d'un mod√®le peut grosso modo √™tre utilis√© comme un proxy de la qualit√© du mod√®le. Il y a plusieurs mesures de la coh√©rence, mais celle impl√©ment√©e par d√©faut est la m√©trique c_v. On mesure la distance entre deux mots qui ont √©t√© assign√©s au m√™me topic. Voir [ce papier](https://liacs.leidenuniv.nl/~spruitmr/pub/2017%20-%20Syed%20Spruit.pdf) pour plus de d√©tails.

```python
benchmark1_lda, bow, benchamrk1_coherence,dictionary = topic_model(data,3)
```
